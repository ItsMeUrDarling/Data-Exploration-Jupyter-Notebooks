## Data Science Job Market Analysis: 
### A Dive into Salaries and Work Trends in 2023

### Introduction
In the rapidly evolving field of data science, understanding the dynamics of the job market is essential for both employers and professionals. This project focuses on analyzing a dataset containing information about data science salaries in 2023, sourced from Kaggle. The data covers various factors like work experience, job title, salary, employment type, remote work ratio, company size and location, and the residence of the employee. Our aim is to explore this dataset to gain insights into various aspects of the data science job market, such as salary distribution, work experience correlation, job diversity, and geographic distribution of opportunities.

#### Import the needed libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

#### Import the data-science-salaries-2023 dataset

```python
df = pd.read_csv('ds_salaries.csv')
```

### 10 Questions to be explored

#### 1. What's the average salary in USD across all jobs?

```python
average_salary_usd = df['salary_in_usd'].mean()
print(f'Average salary in USD: {average_salary_usd}')
```

##### Histogram for distribution of salaries in USD

```python
plt.figure(figsize=(10,6))
sns.histplot(df['salary_in_usd'], kde=True, bins=30)
plt.title('Salary Distribution')
plt.xlabel('Salary in USD')
plt.ylabel('Frequency')
plt.show()
```

##### Boxplot for distribution of salaries in USD

```python
plt.figure(figsize=(10,6))
sns.boxplot(x=df['salary_in_usd'])
plt.title('Salary Distribution')
plt.xlabel('Salary in USD')
plt.show()
```

##### Insights
The average salary across all jobs in the dataset is approximately $137,570.39 USD. This suggests that the average income for data science professionals in this dataset is quite high. The histogram and boxplot will give a better understanding of the salary distribution and will show where most of the salaries lie within this average.

#### 2. How many unique job titles are in the dataset?

```python
unique_job_titles = df['job_title'].nunique()
print(f'Number of unique job titles: {unique_job_titles}')
```

```python
job_titles = df['job_title'].value_counts()
plt.figure(figsize=(10,6))
sns.barplot(x=job_titles.index, y=job_titles.values)
plt.title('Job Titles Distribution')
plt.xlabel('Job Title')
plt.ylabel('Frequency')
plt.xticks(rotation=90)
plt.show()
```

##### Insights
There are 93 unique job titles in the dataset, indicating a variety of roles and positions within the field of data science. These can range from specific roles like "AI Developer" or "Data Analyst" to more general roles like "Data Scientist".

#### 3. What's the correlation between work experience (in years) and salary in USD?

```python
correlation_experience_salary = df['work_year'].corr(df['salary_in_usd'])
print(f'Correlation between work experience and salary: {correlation_experience_salary}')
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x=df['work_year'], y=df['salary_in_usd'])
plt.title('Correlation between work experience and salary')
plt.xlabel('Work Experience (years)')
plt.ylabel('Salary in USD')
plt.show()
```

##### Insights
The correlation between work experience and salary is approximately 0.228. This is a positive correlation, although not very strong, which indicates that as work experience increases, the salary tends to increase as well. This correlation is visually depicted in the scatter plot, where each point represents an observation from the dataset.

#### 4. How many jobs are there for each experience level?

```python
jobs_per_experience_level = df['experience_level'].value_counts()
print(jobs_per_experience_level)
```

```python
experience_level = df['experience_level'].value_counts()
plt.figure(figsize=(10,6))
sns.barplot(x=experience_level.index, y=experience_level.values)
plt.title('Jobs Distribution per Experience Level')
plt.xlabel('Experience Level')
plt.ylabel('Number of Jobs')
plt.show()
```

##### Insights
There are four experience levels represented in the dataset: 'SE' (Senior), 'MI' (Middle), 'EN' (Entry), and 'EX' (Expert). The majority of jobs are for 'SE' level (2516 jobs), followed by 'MI' (805 jobs), 'EN' (320 jobs), and 'EX' (114 jobs). This suggests that there are more job opportunities for senior professionals in the dataset.

#### 5. What is the average salary for each employment type (FT or CT)?

```python
average_salary_per_employment_type = df.groupby('employment_type')['salary_in_usd'].mean()
print(average_salary_per_employment_type)
```

```python
average_salary_per_employment_type = df.groupby('employment_type')['salary_in_usd'].mean()
plt.figure(figsize=(10,6))
sns.barplot(x=average_salary_per_employment_type.index, y=average_salary_per_employment_type.values)
plt.title('Average Salary per Employment Type')
plt.xlabel('Employment Type')
plt.ylabel('Average Salary in USD')
plt.show()
```

##### Insights
The average salary varies significantly by employment type: 

'FT' (Full Time) roles have an average salary of approximately $138,314.20 

'CT' (Contract) roles have an average salary of approximately $113,446.90 

'FL' (Freelance) roles have an average salary of approximately $51,807.80 

'PT' (Part Time) roles have an average salary of approximately $39,533.71. 

This could be due to the different levels of commitment and work hours expected for each type of employment.

#### 6. What's the correlation between remote work ratio and salary in USD?

```python
correlation_remote_ratio_salary = df['remote_ratio'].corr(df['salary_in_usd'])
print(f'Correlation between remote work ratio and salary: {correlation_remote_ratio_salary}')
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x=df['remote_ratio'], y=df['salary_in_usd'])
plt.title('Correlation between Remote Work Ratio and Salary')
plt.xlabel('Remote Work Ratio')
plt.ylabel('Salary in USD')
plt.show()
```

##### Insights
The correlation between the remote work ratio and salary is approximately -0.064. This is a weak negative correlation, which suggests that there's not a strong relationship between these two variables. This implies that the ability to work remotely does not have a significant impact on salary.

#### 7. What's the maximum salary for each job title?

```python
max_salary_per_job_title = df.groupby('job_title')['salary_in_usd'].max()
print(max_salary_per_job_title)
```

```python
max_salary_per_job_title = df.groupby('job_title')['salary_in_usd'].max().nlargest(10)
plt.figure(figsize=(10,6))
sns.barplot(x=max_salary_per_job_title.index, y=max_salary_per_job_title.values)
plt.title('Top 10 Job Titles with the Highest Maximum Salary')
plt.xlabel('Job Title')
plt.ylabel('Maximum Salary in USD')
plt.xticks(rotation=90)
plt.show()
```

##### Insights
The maximum salary varies considerably across the different job titles. For example, the maximum salary for an "AI Scientist" is much higher than for a "3D Computer Vision Researcher". This is likely due to the different skills and expertise required for these roles.

#### 8. How many jobs are there for each company size (S, M, L)?

```python
jobs_per_company_size = df['company_size'].value_counts()
print(jobs_per_company_size)
```

```python
jobs_per_company_size = df['company_size'].value_counts()
plt.figure(figsize=(10,6))
sns.barplot(x=jobs_per_company_size.index, y=jobs_per_company_size.values)
plt.title('Jobs Distribution per Company Size')
plt.xlabel('Company Size')
plt.ylabel('Number of Jobs')
plt.show()
```

##### Insights
Most of the jobs in the dataset are offered by medium-sized ('M') companies (3153 jobs), followed by large-sized ('L') companies (454 jobs), and small-sized ('S') companies (148 jobs). This may reflect the market demand and the ability of different-sized companies to hire data science professionals.

#### 9. What is the distribution of company locations?

```python
company_location_distribution = df['company_location'].value_counts()
print(company_location_distribution)
```

```python
company_location_distribution = df['company_location'].value_counts().nlargest(10)
plt.figure(figsize=(10,6))
sns.barplot(x=company_location_distribution.index, y=company_location_distribution.values)
plt.title('Top 10 Company Locations')
plt.xlabel('Company Location')
plt.ylabel('Number of Jobs')
plt.xticks(rotation=90)
plt.show()
```

##### Insights
The majority of the companies offering jobs are located in the US (3040 jobs), followed by GB (172 jobs), CA (87 jobs), ES (77 jobs), and IN (58 jobs). The other countries offer significantly fewer jobs. The bar chart for the top 10 company locations gives a clearer picture of this distribution.

#### 10. How many job entries are from each employee's residence country?

```python
entries_per_residence_country = df['employee_residence'].value_counts()
print(entries_per_residence_country)
```

```python
job_entries_per_country = df['employee_residence'].value_counts().nlargest(10)
plt.figure(figsize=(10,6))
sns.barplot(x=job_entries_per_country.index, y=job_entries_per_country.values)
plt.title('Top 10 Employee Residences by Job Entries')
plt.xlabel('Employee Residence Country')
plt.ylabel('Number of Job Entries')
plt.xticks(rotation=90)
plt.show()
```

##### Insights
Most of the employees reside in the US (3004 employees), followed by GB (167 employees), CA (85 employees), ES (80 employees), and IN (71 employees). This distribution might be affected by various factors such as job opportunities, immigration policies, and the presence of tech industries in these countries.

### Conclusion

Through this analysis, we have gained several key insights into the state of the data science job market in 2023. The average salary across all data science roles is approximately $137,570, reflecting a lucrative market for professionals. We found a wide range of job titles, indicating the diverse opportunities within the field. The work experience shows a positive correlation with the salary, emphasizing the value of experience in this industry.

The distribution of employment types revealed that full-time roles offer higher average salaries compared to other types of employment. The correlation between remote work ratio and salary was weak, indicating that the ability to work remotely does not significantly impact salary. We also observed that job offerings are predominantly from medium-sized companies and are geographically concentrated in specific countries like the US and the UK.
